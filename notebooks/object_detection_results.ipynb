# Object detection pipeline results
This notebook runs the `detect_objects` pipeline from `src/object_detector.py` and shows intermediate results inline: original, blur, local threshold, cleaned mask, edges, contours overlay, and final annotated boxes.

```python
import sys, os
# ensure project root is on sys.path so `from src...` works when this notebook is opened from the notebooks/ folder
sys.path.insert(0, os.path.abspath('..'))
from src.object_detector import detect_objects
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
res = detect_objects()
print('Available keys:', list(res.keys()))
```

```python
# 1) Original
plt.figure(figsize=(6,4))
plt.imshow(res['original'], cmap='gray')
plt.title('Original')
plt.axis('off')
plt.show()
```

```python
# 2) Blur
plt.figure(figsize=(6,4))
plt.imshow(res['blur'], cmap='gray')
plt.title('Blurred')
plt.axis('off')
plt.show()
```

```python
# 3) Local threshold (display as heatmap)
plt.figure(figsize=(6,4))
plt.imshow(res['local_thresh'], cmap='viridis')
plt.title('Local threshold (float values)')
plt.axis('off')
plt.show()
```

```python
# 4) Cleaned mask
plt.figure(figsize=(6,4))
plt.imshow(res['mask_clean'], cmap='gray')
plt.title('Cleaned mask')
plt.axis('off')
plt.show()
```

```python
# 5) Edges
plt.figure(figsize=(6,4))
plt.imshow(res['edges'], cmap='gray')
plt.title('Edges (Canny)')
plt.axis('off')
plt.show()
```

```python
# 6) Contours overlay
plt.figure(figsize=(6,4))
plt.imshow(cv2.cvtColor(res['img_contours'], cv2.COLOR_BGR2RGB))
plt.title('Contours overlay')
plt.axis('off')
plt.show()
```

```python
# 7) Final annotated boxes
plt.figure(figsize=(6,4))
plt.imshow(cv2.cvtColor(res['img_boxes'], cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
